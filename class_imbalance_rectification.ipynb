# Dealing with class imbalance
Our first models all overfit to the class "no roads", producing all black images as output. This is a classic problem of class imbalance, that we will try to rectify with the help of this article: https://machinelearningmastery.com/tactics-to-combat-imbalanced-classes-in-your-machine-learning-dataset/

## Understanding our imbalance
Let's start by assessing how imbalanced our dataset is

```python
%matplotlib inline
import os,sys
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# Helper functions

def load_image(infilename):
    data = mpimg.imread(infilename)
    return data

def img_float_to_uint8(img):
    rimg = img - np.min(img)
    rimg = (rimg / np.max(rimg) * 255).round().astype(np.uint8)
    return rimg

# Concatenate an image and its groundtruth
def concatenate_images(img, gt_img):
    nChannels = len(gt_img.shape)
    w = gt_img.shape[0]
    h = gt_img.shape[1]
    if nChannels == 3:
        cimg = np.concatenate((img, gt_img), axis=1)
    else:
        gt_img_3c = np.zeros((w, h, 3), dtype=np.uint8)
        gt_img8 = img_float_to_uint8(gt_img)          
        gt_img_3c[:,:,0] = gt_img8
        gt_img_3c[:,:,1] = gt_img8
        gt_img_3c[:,:,2] = gt_img8
        img8 = img_float_to_uint8(img)
        cimg = np.concatenate((img8, gt_img_3c), axis=1)
    return cimg
```

```python
root_dir = "training/"

image_dir = root_dir + "groundtruth/"
files = os.listdir(image_dir)
n = min(100, len(files)) # Load maximum 100 images
print("Loading " + str(n) + " images")
groundtruth_imgs = np.array([load_image(image_dir + files[i]) for i in range(n)])

image_dir = root_dir + "images/"
files = os.listdir(image_dir)
n = min(100, len(files)) # Load maximum 100 images
print("Loading " + str(n) + " images")
sat_imgs = np.array([load_image(image_dir + files[i]) for i in range(n)])
```

```python
def evaluate_balance(imgs, verbose=True):
    no_roads_counter = 0
    roads_counter = 0

    for pixel in imgs.flatten():
        if pixel == 0:
            no_roads_counter += 1
        else:
            roads_counter += 1
    
    if(verbose):
        print("Number of pixels belonging to class \"no roads\": ", no_roads_counter)
        print("Number of pixels belonging to class \"roads\": ", roads_counter)
        print("Class ratio: ", roads_counter/no_roads_counter)
        
    if no_roads_counter == 0:
        no_roads_counter = 1 #this is to avoid division by zero
        
    return roads_counter/no_roads_counter
```

```python
evaluate_balance(groundtruth_imgs)
```

We have a class ratio of around 0.29, which means that we have 3.44 time more "no_roads" pixels than "roads pixels.
Although I expected the imbalance to be far greater, this is probably still enough to overfit on the "no_roads" class, since it will produce 71% of correct guesses, a 71% accuracy.
Part of the problem is that we optimize over this accuracy metric, while it does not reflect well the performance over our dataset. Let's have a look at what the article suggests we can do.

## What can be done?
The article suggests 8 tactics to combat class, let's have a look at each of them and see how they can apply to our problem:

- **Collect more data:** While it is not possible to collect more data (not because it's hard to find satellite images, but because creating the groundtruth images is much too time consuming), since we are dealing with a convolutional problem, we have other techniques to increase our dataset size, such as rotation of images and dividing our images in more, smaller images.
- **Changing the performance metric:** As seen above, accuracy is not a metric that reflects the performance of our model well. The F1 score is a better metric to evaluate our model, but it cannot be used to optimize a model. The article suggests other performance metrics, but it is not said if they can be used as losses to optimize our model, we will look at this later. They are: confusion matrix, precision, recall, kappa and ROC curves.
- **Resampling the dataset:** The idea here is to either add copies of the "roads" class (over-sampling) or delete copies of the "no roads" class (under-sampling) in order to selectively balance our dataset. I think under-sampling combined with the dataset size increase suggested above would work really well to have a big, balanced dataset.
- **Generating synthetic samples:** I don't really get it but I think it's some kind of over-sampling using an algorithm to create new samples of the minority class. It looks like too much of a hassle to use in our case.
- **Use different algorithms:** We are already doing this.
- **Use penalized methods:** Penalized classification imposes an additional cost on the model for making classification mistakes on the minority class during training. These penalties can bias the model to pay more attention to the minority class. If we choose to upsample as suggested above, this would not really be useful.
- **Try a different perspective:** Models that deal specifically with unbalanced classification problems exist. Since we are building our own model, we will not use those, but they could be useful to study/replicate.
- **Get creative:** Great advice, thanks.

I suggest that we increase our dataset size by rotation and division, and then under-sample to get a balanced dataset. I expect any classification model to work better on this balanced dataset. Our goal is to achieve a class ratio closer 1:1 although we don't need an exact ratio of 1:1.

## Increase the dataset size
We currently have 100 400x400 training images, that's 16 million individual pixels to classify.

By dividing, we can get 400 200x200 images.

By rotating in all directions, we can get 1600 200x200 images.

This effectively increases the dataset size by 16 easily, but we can do even better.

We can extract even more sub-images from any image by taking more than 4 200x200 images. This will result in repeated data in the dataset, but this is not really a bad thing in a convolutional problem.

Also it might be easier to undersample later if we work with smaller sub-images than 200x200 pixels, some of those would contain a larger proportion of roads than there would be with a larger image.

We can rotate any image in 4 directions to multiply the dataset size by 4. We could also flip horizontally and vertically to multiply by 3, but in our dataset I think it makes sense to not flip, because most roads follow the same angle, rotation conserves that angle but not flipping.

**Final suggestion:** Divide the training images in multiple 200x200 and 100x100 sub-images. Rotate all images and sub-images in all directions.

### Subsample

```python
def subsample(imgs, groundtruth, subsampling_size, increment_size):
    
    increments = np.linspace(0, len(imgs[0])-subsampling_size, (len(imgs[0])-subsampling_size)/increment_size + 1).astype(int)
    
    subsampled_groundtruth = np.empty((len(imgs)*len(increments)**2, subsampling_size, subsampling_size))
    subsampled_imgs = np.empty((len(imgs)*len(increments)**2, subsampling_size, subsampling_size, 3)) # 3 rgb channels
    
    counter = 0
    
    for index in range(len(imgs)):
        groundtruth_img = groundtruth[index]
        img = imgs[index]
        
        for incrementx in increments:
            for incrementy in increments:
                subsampled_groundtruth[counter] = groundtruth_img[incrementx:incrementx+subsampling_size, incrementy:incrementy+subsampling_size]
                subsampled_imgs[counter] = img[incrementx:incrementx+subsampling_size, incrementy:incrementy+subsampling_size, :]
                counter += 1
                
    return subsampled_imgs, subsampled_groundtruth
```

### Rotate

```python
def rotated_expansion(imgs):
    shape = [imgs.shape[i] for i in range(len(imgs.shape))]
    shape[0] = shape[0]*4 # there will be 4 times as many images after we rotate in each direction
    shape = tuple(shape)
    rotated_imgs = np.empty(shape)
    
    for index in range(len(imgs)):
        img = imgs[index]
        img90 = np.rot90(img)
        img180 = np.rot90(img90)
        img270 = np.rot90(img180)
        
        rotated_imgs[index*4] = img
        rotated_imgs[index*4+1] = img90
        rotated_imgs[index*4+2] = img180
        rotated_imgs[index*4+3] = img270
    
    return rotated_imgs
```

## Under-sampling the new dataset
Now the core of the balancing takes place. We have a lot of individual images and we want to remove images such that we resulting dataset is better balance. Let's begin by having a look at the class balance in each image

```python
ratios = []
imgs = []

for dataset in groundtruth_images:
    for img in dataset:
        ratios.append(evaluate_balance(img, verbose=False))
        imgs.append(img)
```

```python
x = [i for i in range(len(ratios))]
plt.scatter(x, ratios)
```

This distribution doesn't tell us much, except that we have a few very high ratios, probably when the no_roads_counter is at 0. We are looking for ratios around 1, so let's see how many of them we have with a margin of 0.1 (0.9 to 1.1 ratios)

```python
good_ratios = []

for ratio in ratios:
    if ratio > 0.9 and ratio < 1.1:
        good_ratios.append(ratio)

len(good_ratios)
```

That is only 164 images with a very well balanced ratio, so we can add them to the final dataset.

This is not many images, but it will give us a great ratio, let's look at it.

```python
supposedly_balanced_dataset = []


for index in range(len(ratios)):
    ratio = ratios[index]
    if ratio > 0.9 and ratio < 1.1:
        supposedly_balanced_dataset.append(imgs[index])
```

```python
flattened_dataset = []
for img in supposedly_balanced_dataset:
    flattened_dataset.extend(img.flatten())
```

```python
evaluate_balance(np.asarray(flattened_dataset))
```

Awesome ratio! We have less than 2 million pixels to evaluate though, which is a lot less than the initial dataset. Let's see if widening the tolerance a bit still gives us a good ratio.
**Here is an important thought** If we only keep the balanced images in the dataset, the model will surely train well, but it will not perform as well on the real dataset, because it will be optimized for balanced images, when our dataset has unbalanced images. So we will need to add unbalanced images, we can achieve this by adding an image with a counterpart that has the opposite ratio. That way each unbalanced image has an oppositely unbalanced image, and the total balance remains the same.
**Warning** We cannot do this ratio analysis with our current dataset as a whole because the images are of different sizes, and ratio doesn't take the size into account. So we must treat the sizes separately. Let's do this

```python
#function that returns the indices of balanced pairs in the dataset
def select_balanced_pairs(dataset):
    ratios = []
    
    indices_to_keep = []

    # compute the balance for each image of the dataset
    for img in dataset:
        ratios.append(evaluate_balance(img, verbose=False))
    
    for index in range(len(ratios)):
        ratio = ratios[index]
        opposite_ratio = 0.0
        
        if ratio == 0.0:
            continue #don't divide by 0
        else:
            opposite_ratio = 1/ratio
        
        #find an image with the opposite ratio
        for searched_index in range(len(ratios)):
            searched_ratio = ratios[searched_index]
            
            if abs(searched_ratio - opposite_ratio) < 0.05:
                if searched_index not in indices_to_keep: #don't add the same image twice
                    indices_to_keep.append(index)
                    indices_to_keep.append(searched_index)
                    break
    
    ratios_to_keep = [ratios[index] for index in indices_to_keep]
    return indices_to_keep, ratios
```

```python
indices_400 = select_balanced_pairs(groundtruth_imgs_400_rot)
len(indices_400)
```

```python
indices_200, ratios_200 = select_balanced_pairs(groundtruth_imgs_200_rot)
len(indices_200)
```

```python
indices_100 = select_balanced_pairs(groundtruth_imgs_100_rot)
len(indices_100)
```

As expected, it is easier to find balanced pairs in the 100x100 range, we even have no balanced pairs in the 400x400 range. The number of images is a lot smaller but we still have more data than in the original dataset, so far so good! Let's compile those datasets and see if they are really balanced.

```python
def extract_balanced_pairs(dataset, indices):
    extracted_dataset = np.asarray([dataset[index] for index in indices])
    return extracted_dataset
```

```python
balanced_dataset_200 = extract_balanced_pairs(groundtruth_imgs_200_rot, indices_200)
evaluate_balance(balanced_dataset_200)
```

```python
balanced_dataset_100 = extract_balanced_pairs(groundtruth_imgs_100_rot, indices_100)
evaluate_balance(balanced_dataset_100)
```

Awesome! Those numbers are staggering! We have more data and our dataset is very well balanced now! Even with only the 100x100 dataset we have more data than the original dataset, but I think it is better if we train on both.
